# Parsing of training outputs 


```python
import re
import matplotlib.pyplot as plt
```

```python
#first load txt files:
with open('trening_BN.txt') as file:
    file_contents = file.read()
```

```python
print(file_contents[0:100])
```

```python
pattern = "Epoch ([\d]+)/1000\n.*loss: (\d+\.\d+).+val_loss: (\d+\.\d+)"
result = re.findall(pattern, file_contents) 
```

```python
result
```

```python
epoch_nums = [int(i[0]) for i in result]
loss = [float(i[1]) for i in result]
val_loss = [float(i[2]) for i in result]
```

```python
#now plot:
#original
plt.plot(epoch_nums, loss, 'r--')
plt.plot(epoch_nums, val_loss, 'b-')
plt.legend(['Training loss', 'Validation loss'])
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('original')
plt.show()
```

```python
#BN
plt.plot(epoch_nums, loss, 'r--')
plt.plot(epoch_nums, val_loss, 'b-')
plt.legend(['Training loss', 'Validation loss'])
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('original')
plt.show()
```

```python
#Dropout
plt.plot(epoch_nums, loss, 'r--')
plt.plot(epoch_nums, val_loss, 'b-')
plt.legend(['Training loss', 'Validation loss'])
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('original')
plt.show()
```

```python
#Early stopping
plt.plot(epoch_nums, loss, 'r--')
plt.plot(epoch_nums, val_loss, 'b-')
plt.legend(['Training loss', 'Validation loss'])
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('original')
plt.show()
```
