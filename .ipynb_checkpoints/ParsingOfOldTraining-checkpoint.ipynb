# Parsing of training outputs 


```python
import re
import matplotlib.pyplot as plt
```

```python
#first load txt files:
with open('trening_BN.txt') as file:
    file_contents = file.read()
```

```python
print(file_contents[0:100])
```

```python
pattern = "Epoch ([\d]+)/1000\n.*loss: (\d+\.\d+).+val_loss: (\d+\.\d+)"
result = re.findall(pattern, file_contents) 
```

```python
result
```

```python
epoch_nums = [int(i[0]) for i in result]
loss = [float(i[1]) for i in result]
val_loss = [float(i[2]) for i in result]
```

```python
#now plot:
plt.rcParams.update({'font.size': 22})
#original
plt.figure(figsize=(10,7))
plt.plot(loss, linewidth=3)
plt.plot(val_loss, linewidth=3)
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'])
plt.show()
```

```python
#BN
plt.figure(figsize=(10,7))
plt.plot(loss, linewidth=3)
plt.plot(val_loss, linewidth=3)
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'])
plt.show()
```

```python
#Dropout
plt.figure(figsize=(10,7))
plt.plot(loss, linewidth=3)
plt.plot(val_loss, linewidth=3)
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'])
plt.show()
```

```python
#Early stopping
plt.figure(figsize=(10,7))
plt.plot(loss[:253], linewidth=3)
plt.plot(val_loss[:253], linewidth=3)
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'])
plt.show()
#patience=15
#najm. dosezen loss:
miniLoss = min(val_loss[:253])
miniLoss
```
